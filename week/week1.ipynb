```python
from nba_api.stats.endpoints import leaguedashplayerstats
import pandas as pd
```

Game data from past 7 days (10/20 - 10/26)


```python
data = leaguedashplayerstats.LeagueDashPlayerStats(season='2025-26')
players = data.get_data_frames()[0]
players.info
players.columns #player data
```

```python
from nba_api.stats.endpoints import leaguedashteamstats
DATE_FROM = "2025-10-01"   
DATE_TO   = "2025-10-27"   

resp = leaguedashteamstats.LeagueDashTeamStats(
    season='2025-26',
    season_type_all_star='Regular Season',   # or 'Pre Season' if you want preseason
    date_from_nullable=DATE_FROM,
    date_to_nullable=DATE_TO,
    per_mode_detailed='PerGame'              # 'Totals' if you want raw sums
)

teams = resp.get_data_frames()[0]
teams # team data 
```

```python
from nba_api.stats.endpoints import leaguegamelog

gamelog = leaguegamelog.LeagueGameLog(
    season='2025-26',
    season_type_all_star='Regular Season'
)
#game data
games = gamelog.get_data_frames()[0]
games
```

```python
#get data of how the game progressed over time...quarter by quarter stats?
from nba_api.stats.endpoints import playbyplayv2
from time import sleep

game_ids = games["GAME_ID"].astype(str).unique()
all_pbp = []
for gid in game_ids:
    try:
        pbp = playbyplayv2.PlayByPlayV2(game_id=gid)
        df = pbp.get_data_frames()[0]
        df["GAME_ID"] = gid  # keep game id for reference
        all_pbp.append(df)
        sleep(0.6)  
    except Exception as e:
        print(f"Failed {gid}: {e}")

pbp_df = pd.concat(all_pbp, ignore_index=True) if all_pbp else pd.DataFrame()
pbp_df
```

```python

pbp_df.to_csv('playbyplay.csv')
```
