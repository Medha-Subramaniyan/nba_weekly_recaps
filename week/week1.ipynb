```python
#all the games and their scores displayed neatly

#parse through the game data, given each game id
#show the home team, the away team, and the final score 

    #parse the 'matchup'  ..home team -> @ , away team -> vs 

games_unique = games.drop_duplicates(subset=['GAME_ID'])
games_unique
game_report = pd.DataFrame()
```

```python
import numpy as np

# Create game_report from games_unique
game_report = games_unique[['GAME_ID', 'MATCHUP']].copy()

# Parse home and away teams from MATCHUP
# Format: "LAL vs. BOS" = LAL is home, "LAL @ BOS" = BOS is home
is_home_game = game_report['MATCHUP'].str.contains(' vs. ')
teams = game_report['MATCHUP'].str.split(r' vs. | @ ', regex=True, expand=True)

game_report['Home'] = np.where(is_home_game, teams[0], teams[1])
game_report['Away'] = np.where(is_home_game, teams[1], teams[0])
game_report['Matchup'] = game_report['MATCHUP']

# Get final scores by grouping games DataFrame (has both teams' scores)
scores = games.groupby('GAME_ID').agg({
    'TEAM_ABBREVIATION': list,
    'PTS': list
}).reset_index()

# Merge scores with game_report to get final scores
def get_scores(row):
    game_id = row['GAME_ID']
    score_data = scores[scores['GAME_ID'] == game_id]
    if len(score_data) == 0:
        return np.nan, np.nan
    
    teams_list = score_data.iloc[0]['TEAM_ABBREVIATION']
    pts_list = score_data.iloc[0]['PTS']
    
    # Find which score belongs to home and which to away
    home_idx = teams_list.index(row['Home']) if row['Home'] in teams_list else None
    away_idx = teams_list.index(row['Away']) if row['Away'] in teams_list else None
    
    if home_idx is not None and away_idx is not None:
        return pts_list[home_idx], pts_list[away_idx]
    return np.nan, np.nan

game_report[['Home_Score', 'Away_Score']] = game_report.apply(
    lambda row: pd.Series(get_scores(row)), axis=1
)


game_report['Final_Score'] = game_report.apply(
    lambda row: f"{row['Away']} @ {row['Home']}: {int(row['Away_Score'])} - {int(row['Home_Score'])}",
    axis=1
)

# Select and reorder columns
game_report = game_report[['GAME_ID', 'Matchup', 'Home', 'Away', 'Home_Score', 'Away_Score', 'Final_Score']]

# Ensure unique games only (drop duplicates by GAME_ID)
game_report = game_report.drop_duplicates(subset=['GAME_ID'], keep='first')
game_report
```

```python
# Score progression and comeback analysis
print("\n" + "=" * 60)
print("SCORE PROGRESSION & COMEBACK ANALYSIS")
print("=" * 60)

if 'pbp_df' not in locals() or pbp_df.empty:
    print("\nNo play-by-play data available (pbp_df is empty). Run the PBP fetch cell first.")
else:
    import numpy as np
    import pandas as pd

    # Helper: clean score margin to numeric (home perspective)
    def to_margin_numeric(series):
        s = series.fillna(0).replace({'TIE': 0, '': 0})
        return pd.to_numeric(s, errors='coerce').fillna(0).astype(int)

    # Helper: count lead changes (ignore ties and repeated margins with same sign)
    def count_lead_changes(margins: pd.Series) -> int:
        if margins is None or len(margins) == 0:
            return 0
        m = margins.replace(0, np.nan).dropna().astype(int)
        if m.empty:
            return 0
        # Reduce consecutive same-sign runs
        signs = np.sign(m.values)
        # Count sign transitions
        return int(np.sum(signs[1:] * signs[:-1] == -1))

    # Build mapping GAME_ID -> (Home, Away)
    if 'game_report' in locals() and not game_report.empty:
        id_to_home = dict(zip(game_report['GAME_ID'].astype(str), game_report['Home']))
        id_to_away = dict(zip(game_report['GAME_ID'].astype(str), game_report['Away']))
    else:
        # Fallback using games_unique parsing if needed
        if 'games_unique' in locals() and not games_unique.empty:
            tmp = games_unique[['GAME_ID', 'MATCHUP']].copy()
            is_home_game = tmp['MATCHUP'].str.contains(' vs. ')
            teams = tmp['MATCHUP'].str.split(r' vs. | @ ', regex=True, expand=True)
            tmp['Home'] = np.where(is_home_game, teams[0], teams[1])
            tmp['Away'] = np.where(is_home_game, teams[1], teams[0])
            id_to_home = dict(zip(tmp['GAME_ID'].astype(str), tmp['Home']))
            id_to_away = dict(zip(tmp['GAME_ID'].astype(str), tmp['Away']))
        else:
            id_to_home, id_to_away = {}, {}

    summaries = []

    # Work on either close games or all games this week
    target_games = close_games if 'close_games' in locals() and len(close_games) > 0 else game_report

    for gid in target_games['GAME_ID'].astype(str).unique():
        gpbp = pbp_df[pbp_df['GAME_ID'].astype(str) == gid].copy()
        if gpbp.empty:
            continue

        # Use SCOREMARGIN (home perspective) when available
        if 'SCOREMARGIN' in gpbp.columns:
            margin = to_margin_numeric(gpbp['SCOREMARGIN'])
        elif 'SCORE' in gpbp.columns:
            # Parse SCORE like "85 - 82" (visitor - home in NBA API). Compute margin (home - visitor).
            parts = gpbp['SCORE'].str.split(' - ')
            try:
                visitor = pd.to_numeric(parts.str[0], errors='coerce')
                home = pd.to_numeric(parts.str[1], errors='coerce')
                margin = (home - visitor).fillna(method='ffill').fillna(0).astype(int)
            except Exception:
                margin = pd.Series([0] * len(gpbp), index=gpbp.index)
        else:
            margin = pd.Series([0] * len(gpbp), index=gpbp.index)

        largest_home_lead = int(margin.max())
        largest_away_lead = int((-margin).max())
        lead_changes = count_lead_changes(margin)
        times_tied = int((margin == 0).sum())

        # Determine winner from final margin
        final_margin = int(margin.replace(0, method='ffill').iloc[-1]) if len(margin) else 0
        winner = 'Home' if final_margin > 0 else ('Away' if final_margin < 0 else 'Tie')

        # Biggest comeback size (max deficit for the eventual winner)
        if winner == 'Home':
            biggest_comeback = int(max(0, -margin.min()))
        elif winner == 'Away':
            biggest_comeback = int(max(0, margin.max()))
        else:
            biggest_comeback = 0

        summaries.append({
            'GAME_ID': gid,
            'Home': id_to_home.get(gid, ''),
            'Away': id_to_away.get(gid, ''),
            'Largest_Home_Lead': largest_home_lead,
            'Largest_Away_Lead': largest_away_lead,
            'Lead_Changes': lead_changes,
            'Times_Tied': times_tied,
            'Final_Margin': final_margin,
            'Winner': winner,
            'Biggest_Comeback': biggest_comeback
        })

    if len(summaries) == 0:
        print("No summaries could be generated (missing mappings or PBP).")
    else:
        score_summary = pd.DataFrame(summaries)
        # Add final score if available
        if 'game_report' in locals() and not game_report.empty:
            score_summary = score_summary.merge(
                game_report[['GAME_ID', 'Final_Score']], on='GAME_ID', how='left'
            )

        # Show top comebacks
        print("\nTop Comebacks (by points overcome):")
        print(score_summary.sort_values('Biggest_Comeback', ascending=False)
              .head(10)
              [['GAME_ID', 'Away', 'Home', 'Final_Score', 'Winner', 'Biggest_Comeback']]
              .to_string(index=False))

        # Show most volatile games (lead changes)
        print("\nMost Lead Changes:")
        print(score_summary.sort_values('Lead_Changes', ascending=False)
              .head(10)
              [['GAME_ID', 'Away', 'Home', 'Final_Score', 'Lead_Changes', 'Times_Tied']]
              .to_string(index=False))

        # Save for later use
        score_progression_summary = score_summary
        print("\nSaved score_progression_summary DataFrame for further analysis/plotting.")
```

```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Top Comebacks - Individual Game Breakdown

if 'pbp_df' in locals() and not pbp_df.empty:
    
    # Function to analyze score progression for a single game
    def analyze_game_comeback(game_id, pbp_data):
        game_pbp = pbp_data[pbp_data['GAME_ID'] == game_id].copy()
        game_pbp = game_pbp.sort_values('EVENTNUM')
        
        comeback_info = {
            'GAME_ID': game_id,
            'Max_Deficit': 0,
            'Max_Lead': 0,
            'Final_Margin': 0,
            'Quarter_Deficits': {}
        }
        
        # Try to extract score margins
        if 'SCOREMARGIN' in game_pbp.columns:
            margins = game_pbp['SCOREMARGIN'].dropna()
            if len(margins) > 0:
                try:
                    margins_str = margins.astype(str).str.replace('+', '')
                    margins_numeric = pd.to_numeric(margins_str, errors='coerce').dropna()
                    
                    if len(margins_numeric) > 0:
                        comeback_info['Max_Deficit'] = abs(margins_numeric.min())
                        comeback_info['Max_Lead'] = margins_numeric.max()
                        comeback_info['Final_Margin'] = abs(margins_numeric.iloc[-1])
                        
                        # Analyze by quarter
                        if 'PERIOD' in game_pbp.columns:
                            for period in sorted(game_pbp['PERIOD'].unique()):
                                period_pbp = game_pbp[game_pbp['PERIOD'] == period]
                                if 'SCOREMARGIN' in period_pbp.columns:
                                    period_margins = period_pbp['SCOREMARGIN'].dropna()
                                    if len(period_margins) > 0:
                                        try:
                                            period_str = period_margins.astype(str).str.replace('+', '')
                                            period_numeric = pd.to_numeric(period_str, errors='coerce').dropna()
                                            if len(period_numeric) > 0:
                                                comeback_info['Quarter_Deficits'][period] = abs(period_numeric.min())
                                        except:
                                            pass
                except Exception as e:
                    pass
        
        return comeback_info
    
    # Analyze all games
    comeback_analyses = []
    for game_id in pbp_df['GAME_ID'].unique():
        analysis = analyze_game_comeback(game_id, pbp_df)
        if analysis['Max_Deficit'] > 0:
            comeback_analyses.append(analysis)
    
    if len(comeback_analyses) > 0:
        comeback_df = pd.DataFrame(comeback_analyses)
        
        # Merge with game_report
        if 'game_report' in locals() and not game_report.empty:
            comeback_df = comeback_df.merge(
                game_report[['GAME_ID', 'Matchup', 'Home', 'Away', 'Final_Score', 'Point_Diff']],
                on='GAME_ID',
                how='left'
            )
        
        # Sort by comeback magnitude
        top_comebacks = comeback_df.nlargest(10, 'Max_Deficit')
        
        # Create detailed visualization
        fig = plt.figure(figsize=(16, 10))
        gs = fig.add_gridspec(2, 2, hspace=0.3, wspace=0.3)
        fig.suptitle('Top 10 Comebacks - Detailed Analysis', fontsize=16, fontweight='bold')
        
        # 1. Top Comebacks Bar Chart
        ax1 = fig.add_subplot(gs[0, :])
        y_pos = np.arange(len(top_comebacks))
        bars = ax1.barh(y_pos, top_comebacks['Max_Deficit'], color='#FF6B6B', alpha=0.8, edgecolor='black')
        ax1.set_yticks(y_pos)
        
        # Create detailed labels
        labels = []
        for _, row in top_comebacks.iterrows():
            if 'Matchup' in row and pd.notna(row['Matchup']):
                label = f"{row['Matchup']} | Deficit: {row['Max_Deficit']:.1f} pts | Final: {row['Final_Margin']:.1f}"
            elif 'Home' in row and 'Away' in row and pd.notna(row['Home']):
                label = f"{row['Away']} @ {row['Home']} | Deficit: {row['Max_Deficit']:.1f} pts"
            else:
                label = f"Game {str(row['GAME_ID'])[-6:]} | Deficit: {row['Max_Deficit']:.1f} pts"
            labels.append(label[:60])  # Truncate long labels
        
        ax1.set_yticklabels(labels, fontsize=9)
        ax1.set_xlabel('Maximum Deficit Overcome (Points)', fontsize=12, fontweight='bold')
        ax1.set_title('Top 10 Biggest Comebacks', fontsize=14, fontweight='bold')
        ax1.grid(True, alpha=0.3, axis='x')
        
        # Add value labels
        for i, (idx, row) in enumerate(top_comebacks.iterrows()):
            ax1.text(row['Max_Deficit'] + 0.5, i, f"{row['Max_Deficit']:.1f} pts", 
                    va='center', fontsize=10, fontweight='bold')
        
        # 2. Comeback Success Rate (if we have final margin data)
        ax2 = fig.add_subplot(gs[1, 0])
        if 'Final_Margin' in top_comebacks.columns:
            # Categorize outcomes
            wins = top_comebacks[top_comebacks['Final_Margin'] <= 3]  # Close wins
            losses = top_comebacks[top_comebacks['Final_Margin'] > 3]  # Lost despite comeback
            
            categories = ['Successful\nComeback Win', 'Comeback\nbut Lost']
            counts = [len(wins), len(losses)]
            colors = ['#4ECDC4', '#FF6B6B']
            
            bars2 = ax2.bar(categories, counts, color=colors, alpha=0.8, edgecolor='black')
            ax2.set_ylabel('Number of Games', fontsize=11, fontweight='bold')
            ax2.set_title('Comeback Outcomes', fontsize=12, fontweight='bold')
            ax2.grid(True, alpha=0.3, axis='y')
            
            # Add value labels
            for bar in bars2:
                height = bar.get_height()
                ax2.text(bar.get_x() + bar.get_width()/2., height,
                        f'{int(height)}', ha='center', va='bottom', fontsize=11, fontweight='bold')
        
        # 3. Deficit vs Final Margin Scatter
        ax3 = fig.add_subplot(gs[1, 1])
        scatter = ax3.scatter(top_comebacks['Max_Deficit'], top_comebacks['Final_Margin'],
                             s=150, alpha=0.7, c=top_comebacks['Max_Deficit'], cmap='RdYlGn',
                             edgecolors='black', linewidth=2)
        ax3.set_xlabel('Maximum Deficit Overcome', fontsize=11, fontweight='bold')
        ax3.set_ylabel('Final Margin', fontsize=11, fontweight='bold')
        ax3.set_title('Comeback Magnitude vs Final Outcome', fontsize=12, fontweight='bold')
        ax3.axhline(y=0, color='gray', linestyle='--', linewidth=1, alpha=0.5)
        ax3.grid(True, alpha=0.3)
        plt.colorbar(scatter, ax=ax3, label='Deficit (pts)')
        
        # Annotate top 3 comebacks
        for i, (idx, row) in enumerate(top_comebacks.head(3).iterrows()):
            if 'Home' in row and pd.notna(row['Home']):
                label = f"{row['Away']} @ {row['Home']}"[:15]
                ax3.annotate(label, (row['Max_Deficit'], row['Final_Margin']),
                            fontsize=8, alpha=0.8, xytext=(5, 5), textcoords='offset points')
        
        plt.tight_layout()
        plt.show()
        
        # Print summary table
        print("\n" + "="*70)
        print("TOP 10 COMEBACKS SUMMARY")
        print("="*70)
        display_cols = ['GAME_ID', 'Matchup', 'Max_Deficit', 'Final_Margin'] if 'Matchup' in top_comebacks.columns else ['GAME_ID', 'Max_Deficit', 'Final_Margin']
        print(top_comebacks[display_cols].to_string(index=False))
        print("="*70)
        
    else:
        print("No comeback data extracted from play-by-play data.")
        print("Note: This requires SCOREMARGIN column in play-by-play data.")

elif 'game_report' in locals() and not game_report.empty:
    # Fallback: Show close games that could have been comebacks
    print("Using game report data to identify potential comebacks...")
    close_games = game_report[game_report['Point_Diff'] <= 5].copy()
    
    if len(close_games) > 0:
        fig, ax = plt.subplots(figsize=(14, 8))
        
        # Sort by point differential
        close_games_sorted = close_games.sort_values('Point_Diff')
        
        y_pos = np.arange(len(close_games_sorted))
        bars = ax.barh(y_pos, close_games_sorted['Point_Diff'], color='#FF6B6B', alpha=0.8)
        ax.set_yticks(y_pos)
        ax.set_yticklabels([f"{row['Matchup']}"[:40] for _, row in close_games_sorted.iterrows()], fontsize=9)
        ax.set_xlabel('Point Differential', fontsize=12, fontweight='bold')
        ax.set_title('Close Games (Potential Comebacks) - Ordered by Competitiveness', fontsize=14, fontweight='bold')
        ax.grid(True, alpha=0.3, axis='x')
        
        for i, (idx, row) in enumerate(close_games_sorted.iterrows()):
            ax.text(row['Point_Diff'] + 0.1, i, f"{row['Point_Diff']:.0f}", 
                   va='center', fontsize=9, fontweight='bold')
        
        plt.tight_layout()
        plt.show()
        
        print("\n" + "="*70)
        print("CLOSE GAMES (POTENTIAL COMEBACKS)")
        print("="*70)
        print(close_games_sorted[['Matchup', 'Final_Score', 'Point_Diff']].to_string(index=False))
        print("="*70)
    else:
        print("No close games found for comeback visualization.")

else:
    print("No data available for comeback analysis. Need play-by-play or game report data.")

```
