```python
#all the games and their scores displayed neatly

#parse through the game data, given each game id
#show the home team, the away team, and the final score 

    #parse the 'matchup'  ..home team -> @ , away team -> vs 

games_unique = games.drop_duplicates(subset=['GAME_ID'])
games_unique
game_report = pd.DataFrame()
```

```python
import numpy as np

# Create game_report from games_unique
game_report = games_unique[['GAME_ID', 'MATCHUP']].copy()

# Parse home and away teams from MATCHUP
# Format: "LAL vs. BOS" = LAL is home, "LAL @ BOS" = BOS is home
is_home_game = game_report['MATCHUP'].str.contains(' vs. ')
teams = game_report['MATCHUP'].str.split(r' vs. | @ ', regex=True, expand=True)

game_report['Home'] = np.where(is_home_game, teams[0], teams[1])
game_report['Away'] = np.where(is_home_game, teams[1], teams[0])
game_report['Matchup'] = game_report['MATCHUP']

# Get final scores by grouping games DataFrame (has both teams' scores)
scores = games.groupby('GAME_ID').agg({
    'TEAM_ABBREVIATION': list,
    'PTS': list
}).reset_index()

# Merge scores with game_report to get final scores
def get_scores(row):
    game_id = row['GAME_ID']
    score_data = scores[scores['GAME_ID'] == game_id]
    if len(score_data) == 0:
        return np.nan, np.nan
    
    teams_list = score_data.iloc[0]['TEAM_ABBREVIATION']
    pts_list = score_data.iloc[0]['PTS']
    
    # Find which score belongs to home and which to away
    home_idx = teams_list.index(row['Home']) if row['Home'] in teams_list else None
    away_idx = teams_list.index(row['Away']) if row['Away'] in teams_list else None
    
    if home_idx is not None and away_idx is not None:
        return pts_list[home_idx], pts_list[away_idx]
    return np.nan, np.nan

game_report[['Home_Score', 'Away_Score']] = game_report.apply(
    lambda row: pd.Series(get_scores(row)), axis=1
)


game_report['Final_Score'] = game_report.apply(
    lambda row: f"{row['Away']} @ {row['Home']}: {int(row['Away_Score'])} - {int(row['Home_Score'])}",
    axis=1
)

# Select and reorder columns
game_report = game_report[['GAME_ID', 'Matchup', 'Home', 'Away', 'Home_Score', 'Away_Score', 'Final_Score']]

# Ensure unique games only (drop duplicates by GAME_ID)
game_report = game_report.drop_duplicates(subset=['GAME_ID'], keep='first')
game_report
```

```python
# Question 2: Close games analysis
print("=" * 60)
print("CLOSE GAMES ANALYSIS")
print("=" * 60)

# Define close game as point differential <= 5
CLOSE_GAME_THRESHOLD = 5
close_games = game_report[game_report['Point_Diff'] <= CLOSE_GAME_THRESHOLD].copy()

if len(close_games) > 0:
    print(f"\nClose Games (Point Differential ≤ {CLOSE_GAME_THRESHOLD}): {len(close_games)}")
    print("\nClose Games Details:")
    print(close_games[['Matchup', 'Final_Score', 'Point_Diff']].to_string(index=False))
    
```

```python
# Score progression and comeback analysis
print("\n" + "=" * 60)
print("SCORE PROGRESSION & COMEBACK ANALYSIS")
print("=" * 60)

if 'pbp_df' not in locals() or pbp_df.empty:
    print("\nNo play-by-play data available (pbp_df is empty). Run the PBP fetch cell first.")
else:
    import numpy as np
    import pandas as pd

    # Helper: clean score margin to numeric (home perspective)
    def to_margin_numeric(series):
        s = series.fillna(0).replace({'TIE': 0, '': 0})
        return pd.to_numeric(s, errors='coerce').fillna(0).astype(int)

    # Helper: count lead changes (ignore ties and repeated margins with same sign)
    def count_lead_changes(margins: pd.Series) -> int:
        if margins is None or len(margins) == 0:
            return 0
        m = margins.replace(0, np.nan).dropna().astype(int)
        if m.empty:
            return 0
        # Reduce consecutive same-sign runs
        signs = np.sign(m.values)
        # Count sign transitions
        return int(np.sum(signs[1:] * signs[:-1] == -1))

    # Build mapping GAME_ID -> (Home, Away)
    if 'game_report' in locals() and not game_report.empty:
        id_to_home = dict(zip(game_report['GAME_ID'].astype(str), game_report['Home']))
        id_to_away = dict(zip(game_report['GAME_ID'].astype(str), game_report['Away']))
    else:
        # Fallback using games_unique parsing if needed
        if 'games_unique' in locals() and not games_unique.empty:
            tmp = games_unique[['GAME_ID', 'MATCHUP']].copy()
            is_home_game = tmp['MATCHUP'].str.contains(' vs. ')
            teams = tmp['MATCHUP'].str.split(r' vs. | @ ', regex=True, expand=True)
            tmp['Home'] = np.where(is_home_game, teams[0], teams[1])
            tmp['Away'] = np.where(is_home_game, teams[1], teams[0])
            id_to_home = dict(zip(tmp['GAME_ID'].astype(str), tmp['Home']))
            id_to_away = dict(zip(tmp['GAME_ID'].astype(str), tmp['Away']))
        else:
            id_to_home, id_to_away = {}, {}

    summaries = []

    # Work on either close games or all games this week
    target_games = close_games if 'close_games' in locals() and len(close_games) > 0 else game_report

    for gid in target_games['GAME_ID'].astype(str).unique():
        gpbp = pbp_df[pbp_df['GAME_ID'].astype(str) == gid].copy()
        if gpbp.empty:
            continue

        # Use SCOREMARGIN (home perspective) when available
        if 'SCOREMARGIN' in gpbp.columns:
            margin = to_margin_numeric(gpbp['SCOREMARGIN'])
        elif 'SCORE' in gpbp.columns:
            # Parse SCORE like "85 - 82" (visitor - home in NBA API). Compute margin (home - visitor).
            parts = gpbp['SCORE'].str.split(' - ')
            try:
                visitor = pd.to_numeric(parts.str[0], errors='coerce')
                home = pd.to_numeric(parts.str[1], errors='coerce')
                margin = (home - visitor).fillna(method='ffill').fillna(0).astype(int)
            except Exception:
                margin = pd.Series([0] * len(gpbp), index=gpbp.index)
        else:
            margin = pd.Series([0] * len(gpbp), index=gpbp.index)

        largest_home_lead = int(margin.max())
        largest_away_lead = int((-margin).max())
        lead_changes = count_lead_changes(margin)
        times_tied = int((margin == 0).sum())

        # Determine winner from final margin
        final_margin = int(margin.replace(0, method='ffill').iloc[-1]) if len(margin) else 0
        winner = 'Home' if final_margin > 0 else ('Away' if final_margin < 0 else 'Tie')

        # Biggest comeback size (max deficit for the eventual winner)
        if winner == 'Home':
            biggest_comeback = int(max(0, -margin.min()))
        elif winner == 'Away':
            biggest_comeback = int(max(0, margin.max()))
        else:
            biggest_comeback = 0

        summaries.append({
            'GAME_ID': gid,
            'Home': id_to_home.get(gid, ''),
            'Away': id_to_away.get(gid, ''),
            'Largest_Home_Lead': largest_home_lead,
            'Largest_Away_Lead': largest_away_lead,
            'Lead_Changes': lead_changes,
            'Times_Tied': times_tied,
            'Final_Margin': final_margin,
            'Winner': winner,
            'Biggest_Comeback': biggest_comeback
        })

    if len(summaries) == 0:
        print("No summaries could be generated (missing mappings or PBP).")
    else:
        score_summary = pd.DataFrame(summaries)
        # Add final score if available
        if 'game_report' in locals() and not game_report.empty:
            score_summary = score_summary.merge(
                game_report[['GAME_ID', 'Final_Score']], on='GAME_ID', how='left'
            )

        # Show top comebacks
        print("\nTop Comebacks (by points overcome):")
        print(score_summary.sort_values('Biggest_Comeback', ascending=False)
              .head(10)
              [['GAME_ID', 'Away', 'Home', 'Final_Score', 'Winner', 'Biggest_Comeback']]
              .to_string(index=False))

        # Show most volatile games (lead changes)
        print("\nMost Lead Changes:")
        print(score_summary.sort_values('Lead_Changes', ascending=False)
              .head(10)
              [['GAME_ID', 'Away', 'Home', 'Final_Score', 'Lead_Changes', 'Times_Tied']]
              .to_string(index=False))

        # Save for later use
        score_progression_summary = score_summary
        print("\nSaved score_progression_summary DataFrame for further analysis/plotting.")
```

```python
# Question 3: Standout players and outlier performances
print("=" * 60)
print("STANDOUT PLAYERS & OUTLIER PERFORMANCES")
print("=" * 60)

# Filter player stats for the week's games
# Check if player_game_stats exists (should be created by the player game log fetching cell)
if 'player_game_stats' not in locals() or player_game_stats.empty:
    print("\nERROR: Player game stats not available.")
    print("Please run the 'Fetch Player Game Logs for This Week' cell first.")
    player_totals = pd.DataFrame()
else:
    # Use player_game_stats directly (already filtered for this week's games)
    game_id_col = 'Game_ID' if 'Game_ID' in player_game_stats.columns else 'GAME_ID'
    team_col = 'TEAM_ABBREVIATION' if 'TEAM_ABBREVIATION' in player_game_stats.columns else None
    
    if 'PLAYER_NAME' not in player_game_stats.columns:
        print(f"\nERROR: PLAYER_NAME column not found.")
        print(f"Available columns: {list(player_game_stats.columns)}")
        player_totals = pd.DataFrame()
    else:
        groupby_cols = ['PLAYER_NAME']
        if team_col:
            groupby_cols.append(team_col)
        
        # Build agg_dict with only columns that exist
        agg_dict = {}
        stat_cols = ['PTS', 'AST', 'REB', 'STL', 'BLK', 'FG3M', 'MIN']
        for col in stat_cols:
            if col in player_game_stats.columns:
                agg_dict[col] = 'sum'
        
        if game_id_col in player_game_stats.columns:
            agg_dict[game_id_col] = 'count'
        
        if len(agg_dict) > 0:
            player_totals = player_game_stats.groupby(groupby_cols).agg(agg_dict).rename(columns={game_id_col: 'Games_Played'}).reset_index()
            
            # Calculate per game averages (only if we have games played)
            if 'Games_Played' in player_totals.columns:
                if 'PTS' in player_totals.columns:
                    player_totals['PPG'] = player_totals['PTS'] / player_totals['Games_Played']
                if 'AST' in player_totals.columns:
                    player_totals['APG'] = player_totals['AST'] / player_totals['Games_Played']
                if 'REB' in player_totals.columns:
                    player_totals['RPG'] = player_totals['REB'] / player_totals['Games_Played']
                
                # Display results
                if 'PTS' in player_totals.columns and 'PPG' in player_totals.columns:
                    print("\nTop 10 Scorers (Total Points):")
                    display_cols = ['PLAYER_NAME']
                    if team_col:
                        display_cols.append(team_col)
                    display_cols.extend(['PTS', 'PPG', 'Games_Played'])
                    print(player_totals.nlargest(10, 'PTS')[display_cols].to_string(index=False))
                    
                    print("\nTop 10 Scorers (Per Game Average):")
                    display_cols = ['PLAYER_NAME']
                    if team_col:
                        display_cols.append(team_col)
                    display_cols.extend(['PPG', 'PTS', 'Games_Played'])
                    print(player_totals.nlargest(10, 'PPG')[display_cols].to_string(index=False))
                
                if 'AST' in player_totals.columns and 'APG' in player_totals.columns:
                    print("\nTop Assist Leaders:")
                    display_cols = ['PLAYER_NAME']
                    if team_col:
                        display_cols.append(team_col)
                    display_cols.extend(['AST', 'APG', 'Games_Played'])
                    print(player_totals.nlargest(10, 'AST')[display_cols].to_string(index=False))
                
                if 'REB' in player_totals.columns and 'RPG' in player_totals.columns:
                    print("\nTop Rebound Leaders:")
                    display_cols = ['PLAYER_NAME']
                    if team_col:
                        display_cols.append(team_col)
                    display_cols.extend(['REB', 'RPG', 'Games_Played'])
                    print(player_totals.nlargest(10, 'REB')[display_cols].to_string(index=False))
                
                if 'FG3M' in player_totals.columns:
                    print("\nTop 3-Point Shooters:")
                    display_cols = ['PLAYER_NAME']
                    if team_col:
                        display_cols.append(team_col)
                    display_cols.extend(['FG3M', 'Games_Played'])
                    print(player_totals.nlargest(10, 'FG3M')[display_cols].to_string(index=False))
            else:
                print("\nNo valid columns found for aggregation")
                player_totals = pd.DataFrame()
        else:
            print("\nNo valid stat columns found in player_game_stats")
            player_totals = pd.DataFrame()
```
